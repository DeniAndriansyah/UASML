```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/wine-quality/winequalityN.csv')
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.head(1000)
```

```python
df['goodquality'] = [1 if x >= 7 else 0 for x in df['quality']]
```

```python
df = df.rename(columns={
     'fixed acidity': 'Fixed_Acidity',
    'volatile acidity': 'Volatile_Acidity',
    'citric acid': 'Citric_Acid',
    'residual sugar': 'Residual_Sugar',
    'free sulfur dioxide': 'Free_Sulfur_Dioxide',
    'total sulfur dioxide': 'Total_Sulfur_Dioxide',
    'density': 'Density',
    'pH': 'pH',
    'sulphates': 'Sulphates',
    'alcohol': 'Alcohol'
})
```

```python
df
```

```python
df['goodquality'].value_counts()
```

```python
df['quality'].value_counts()
```

```python
x=df.drop(['quality','type','goodquality'] , axis = 1)
y=df['goodquality']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = tf.keras.Sequential([
    keras.layers.Flatten(input_shape=(11, )),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
x_test.columns
```

```python
model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test), batch_size=10)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)

print(f'Loss: {loss}')
print(f'Accuracy: {accuracy}')
```
